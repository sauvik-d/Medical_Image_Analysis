```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ['Decision Tree', 'Random Forest', 'SVM', 'ANN', 'CNN', 'DenseNet121']
```

```python
precision_without_dcgan = [0.7539, 0.7882, 0.8338, 0.8123, 0.8102, 0.7647]
precision_with_dcgan    = [0.7439, 0.7858, 0.8333, 0.8135, 0.8472, 0.7613]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width/2, precision_without_dcgan, width, label='Without DCGAN', color='skyblue')
ax.bar(x + width/2, precision_with_dcgan, width, label='With DCGAN', color='orange')
ax.set_ylabel('Precision')
ax.set_title('Precision Comparison (With vs Without DCGAN)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0, 1)
ax.legend()

for bar in ax.containers:
    ax.bar_label(bar, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.savefig('precision_new.png', dpi=300)
plt.show()
```

```python
recall_without_dcgan = [0.7324, 0.7388, 0.7933, 0.7484, 0.7372, 1]
recall_with_dcgan    = [0.7372, 0.7500, 0.7885, 0.7804, 0.8317, 0.9974]

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width/2, recall_without_dcgan, width, label='Without DCGAN', color='lightgreen')
ax.bar(x + width/2, recall_with_dcgan, width, label='With DCGAN', color='green')
ax.set_ylabel('Recall')
ax.set_title('Recall Comparison (With vs Without DCGAN)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0, 1)
ax.legend()

for bar in ax.containers:
    ax.bar_label(bar, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.savefig('recall_new.png', dpi=300)
plt.show()
```

```python
f1_without_dcgan = [0.7004, 0.6987, 0.7711, 0.7079, 0.6902, 0.8667]
f1_with_dcgan    = [0.7151, 0.7182, 0.7642, 0.7575, 0.8220, 0.8635]

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width/2, f1_without_dcgan, width, label='Without DCGAN', color='steelblue')
ax.bar(x + width/2, f1_with_dcgan, width, label='With DCGAN', color='darkorange')
ax.set_ylabel('F1 Score')
ax.set_title('F1 Score Comparison (With vs Without DCGAN)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0, 1)
ax.legend()

for bar in ax.containers:
    ax.bar_label(bar, fmt='%.3f', padding=3, fontsize=8)

plt.tight_layout()
plt.savefig('f1_new.png', dpi=300)
plt.show()
```

```python
roc_auc_without_dcgan = [0.6594, 0.6577, 0.7278, 0.8418, 0.8031, 0.9074]
roc_auc_with_dcgan    = [0.6752, 0.6761, 0.7205, 0.8569, 0.8728, 0.9568]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, roc_auc_without_dcgan, marker='o', linestyle='-', color='mediumpurple', label='Without DCGAN')
ax.plot(x, roc_auc_with_dcgan, marker='s', linestyle='--', color='seagreen', label='With DCGAN')

ax.set_ylabel('ROC-AUC')
ax.set_title('ROC-AUC Comparison (With vs Without DCGAN)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0.6, 1.0)
ax.legend()

for i in range(len(x)):
    ax.text(x[i], roc_auc_without_dcgan[i] - 0.015, f'{roc_auc_without_dcgan[i]:.4f}', 
            ha='center', va='top', fontsize=8, color='mediumpurple')
    ax.text(x[i], roc_auc_with_dcgan[i] + 0.015, f'{roc_auc_with_dcgan[i]:.4f}', 
            ha='center', va='bottom', fontsize=8, color='seagreen')

plt.grid(True)
plt.tight_layout()
plt.savefig('roc_new.png', dpi=300)
plt.show()
```
