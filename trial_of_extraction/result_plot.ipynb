```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ['Decision Tree', 'Random Forest', 'SVM', 'ANN', 'CNN', 'DenseNet121']

# Precision values: [without DCGAN, with DCGAN]
precision_without_dcgan = [0.7444, 0.8243, 0.8338, 0.8106, 0.7944, 0.7553]
precision_with_dcgan    = [0.7747, 0.8394, 0.8333, 0.8124, 0.8226, 0.7514]

# X-axis positions
x = np.arange(len(models))
width = 0.35  # width of the bars
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, precision_without_dcgan, width, label='Without DCGAN', color='skyblue')
bars2 = ax.bar(x + width/2, precision_with_dcgan, width, label='With DCGAN', color='orange')

# Labels and titles
ax.set_ylabel('Precision')
ax.set_title('Precision Comparison (With vs Without DCGAN)')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=0, ha='center')
ax.set_ylim(0, 1)
ax.legend()

# Annotate bars with values
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 4),
                textcoords='offset points',
                ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('precision_plot.png')
plt.show()
```

```python
recall_without_dcgan = [0.7260, 0.7628, 0.7933, 0.7821, 0.7324, 0.9974]
recall_with_dcgan    = [0.7628, 0.8109, 0.7885, 0.7869, 0.7837, 1.0000]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, recall_without_dcgan, width, label='Without DCGAN', color='lightgreen')
bars2 = ax.bar(x + width/2, recall_with_dcgan, width, label='With DCGAN', color='darkgreen')

ax.set_ylabel('Recall')
ax.set_title('Recall Comparison (With vs Without DCGAN)')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=0, ha='center')
ax.set_ylim(0, 1.1)
ax.legend()

for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 4),
                textcoords='offset points',
                ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('recall_plot.png')
plt.show()
```

```python
# F1-Score values
f1_without_dcgan = [0.6933, 0.7277, 0.7711, 0.7609, 0.6862, 0.8597]
f1_with_dcgan    = [0.7441, 0.7950, 0.7642, 0.7676, 0.7598, 0.8581]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, f1_without_dcgan, width, label='Without DCGAN', color='steelblue')
bars2 = ax.bar(x + width/2, f1_with_dcgan, width, label='With DCGAN', color='salmon')

ax.set_ylabel('F1-Score')
ax.set_title('F1-Score Comparison (With vs Without DCGAN)')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=0, ha='center')
ax.set_ylim(0, 1.1)
ax.legend()

for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 4),
                textcoords='offset points',
                ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('f1_plot.png')
plt.show()
```

```python
roc_auc_without_dcgan = [0.6526, 0.6846, 0.7278, 0.8691, 0.8547, 0.9163]
roc_auc_with_dcgan    = [0.7043, 0.7538, 0.7205, 0.8666, 0.8062, 0.9542]

x = np.arange(len(models))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, roc_auc_without_dcgan, marker='o', linestyle='-', label='Without DCGAN', color='mediumpurple')
plt.plot(x, roc_auc_with_dcgan, marker='s', linestyle='--', label='With DCGAN', color='seagreen')

# Labels and formatting
plt.xticks(x, models, rotation=0)
plt.ylim(0.6, 1.0)
plt.ylabel('ROC-AUC Score')
plt.title('ROC-AUC Score Comparison (With vs Without DCGAN)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Annotate values
for i, (wo, wd) in enumerate(zip(roc_auc_without_dcgan, roc_auc_with_dcgan)):
    plt.text(x[i], wo - 0.02, f'{wo:.2f}', ha='center', va='top', fontsize=8, color='steelblue')
    plt.text(x[i], wd + 0.02, f'{wd:.2f}', ha='center', va='bottom', fontsize=8, color='darkorange')
plt.savefig('rocauc_plot.png')
plt.show()
```
